# WordCloud
A word cloud is a visual representation of text data where the size of each word corresponds to its frequency or importance in the given text. It provides a quick and intuitive way to understand the most common or prominent words in a body of text.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text=("Datascience Python Python Python Matplotlib Datascience Matplotlib Education Seaborn Network Plot Datascience Violin Chart Pandas Education Datascience Wordcloud Spider Radar Education Parrallel Education Alpha Color Brewer Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble")
```

```python
text
```

```python
#create wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)
```

```python
#visualize generated image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```
